## **Intro to Agentic AI**

**Background**

Before `Agents`, we had models that were static, responding to prompts without any memory or ability to perform tasks autonomously.

They did not have memory, so they couldn't remember past interactions or learn from them. They also couldn't perform tasks on their own; they needed explicit instructions for every action.

They could only respond to prompts, not decide on their own what to do next.

**What are Agents?**

Agents are a new paradigm in AI that combines the capabilities of language models with memory and autonomy. They can remember past interactions, learn from them, and perform tasks on their own without needing explicit instructions for every action.

Agents can decide on their own what to do next based on their goals and the information they have. They can also interact with external tools and resources to accomplish tasks.

**Key Features of Agents**

1. **Memory**: Agents can remember past interactions and learn from them, allowing for more personalized and context-aware responses.

2. **Autonomy**: Agents can perform tasks on their own without needing explicit instructions for every action. They can decide on their own what to do next based on their goals and the information they have.

3. **Tool Use**: Agents can interact with external tools and resources to accomplish tasks, such as accessing databases, using APIs, or controlling smart devices.

4. **Goal-Oriented**: Agents can set and pursue goals, allowing them to work towards specific outcomes rather than just responding to prompts.

<hr>

Long story short, agents are group of models that talk to each other, have memory, and can perform tasks on their own. They are a powerful new paradigm in AI that has the potential to revolutionize how we interact with technology and solve complex problems.

If we just try to connect multiple `LLMs` together to form an agent, we will run into issues with memory and coordination. We need a way to manage the interactions between the models and ensure that they can work together effectively.

This is where `LangGraph` and `LangChain` come in. They provide a framework for building agents that can manage memory, coordinate interactions between models, and perform tasks autonomously.

## **LangChain**

`LangChain` is a framework for building agents that can manage memory, coordinate interactions between models, and perform tasks autonomously. It provides modular building blocks that let you create LLM based workflows, agents, and applications.


`LangChain` contains:

**Model** : This is the core component of `LangChain`. It provides a simple interface for interacting with language models. You can use it to generate text, answer questions, and perform other language-related tasks.

**Prompts** : This component allows you to create and manage prompts for your language models. You can use it to create templates, manage variables, and generate prompts dynamically.

**Retrievers** : This component provides tools for retrieving information from external sources, such as databases or APIs. You can use it to access information that your language models may not have in their training data.

But the biggest offering of `LangChain` is `Chains`.

<hr>

### **Applications of `LangChain`**

1. **Chatbots**: You can use `LangChain` to build chatbots that can have more natural and engaging conversations with users. The memory and autonomy features of `LangChain` allow chatbots to remember past interactions and provide more personalized responses.

2. **RAG (Retrieval Augmented Generation)**: You can use `LangChain` to build RAG systems that can retrieve information from external sources and use it to generate more accurate and relevant responses.

3. **Basic Level Agents**: You can use `LangChain` to build basic level agents that can perform simple tasks autonomously, such as scheduling appointments or sending emails. `LangChain` provides `Tools` that allow agents to interact with external resources and perform tasks on their own. For example, an agent could use a `calendar API` to schedule appointments or an email API to send emails without needing explicit instructions for every action.
