## **Intro to Agentic AI**

**Background**

Before `Agents`, we had models that were static, responding to prompts without any memory or ability to perform tasks autonomously.

They did not have memory, so they couldn't remember past interactions or learn from them. They also couldn't perform tasks on their own; they needed explicit instructions for every action.

They could only respond to prompts, not decide on their own what to do next.

**What are Agents?**

Agents are a new paradigm in AI that combines the capabilities of language models with memory and autonomy. They can remember past interactions, learn from them, and perform tasks on their own without needing explicit instructions for every action.

Agents can decide on their own what to do next based on their goals and the information they have. They can also interact with external tools and resources to accomplish tasks.

**Key Features of Agents**

### **Memory**

Agents can remember past interactions and learn from them, allowing for more personalized and context-aware responses.

### **Autonomy**

Agents can perform tasks on their own without needing explicit instructions for every action. They can decide on their own what to do next based on their goals and the information they have.

Autonomy can be controlled using: 

**Permission Scope**: Limit what tools or actions the agent can perform independently.

**Human-in-the-Loop**: Allow human oversight and intervention when necessary.

**Override Controls**: Provide mechanisms for users to override the agent's decisions or actions. Allow users to stop, pause or change the agent's behavior if needed.

**Guardrails / Policies**: Define hard rules of ethical boundaries that the agent cannot cross, ensuring it operates within acceptable limits.

### **Tool Use**

Agents can interact with external tools and resources to accomplish tasks, such as accessing databases, using APIs, or controlling smart devices.

### **Goal-Oriented**

Agents can set and pursue goals, allowing them to work towards specific outcomes rather than just responding to prompts.

### **Planning**

Agents can plan their actions to achieve their goals, breaking down complex tasks into smaller steps and executing them in a logical sequence.

### **Reasoning**

Agents can reason about their actions and the information they have, allowing them to make informed decisions and solve problems effectively.

For complex tasks, it needs to reason well about the task so that it can break it down into smaller steps and execute them in a logical sequence. This requires the agent to have a good understanding of the task and the information it has, as well as the ability to plan and make decisions based on that information.

### **Adaptability**

Agents can adapt to new information and changing circumstances, allowing them to adjust their behavior and strategies as needed to achieve their goals.

### **Context Awareness**

Agents can understand and utilize context from past interactions, current environment, and external information to make informed decisions and provide relevant responses.

Context can be from: 

- Goal

- Progress till now

- Environment state

- Tool Responses

- User specific preferences or history

- Policy or Guardrails

> Context awareness is implemented through memory

## **Components of Agents**

Agents are made up of several key components that work together to enable their functionality:

### **Language Model**

The language model is the core component of an agent, responsible for understanding and generating human-like text. It processes input from users and generates responses based on its training data and the context it has.

### **Orchestrator**

The orchestrator is responsible for managing the flow of information and actions within the agent. It coordinates the interactions between the language model, memory, tools, and other components to ensure that the agent operates effectively and efficiently.

It handles the decision-making process, determining which tools to use, when to access memory, and how to respond to user inputs based on the agent's goals and the current context.

For example, most popular agent frameworks like `LangSmith` and `AgentGPT` have an orchestrator component that manages the flow of information and actions within the agent.

### **Tools**

Tools are external resources or functionalities that the agent can utilize to accomplish tasks. This can include APIs, databases, smart devices, or any other resource that can be accessed programmatically.

Agents can use tools to gather information, perform actions, or interact with the environment in ways that go beyond just generating text responses.

An agent might call `MCP` server to get the `figma` design file, or call `Google Search API` to get the latest information on a topic, or call `Python REPL` to execute code and get results.

### **Memory**

Memory allows the agent to store and recall information from past interactions. This enables the agent to provide more personalized and context-aware responses, as well as to learn from previous experiences.

Memory can be short-term, such as remembering the current conversation, or long-term, such as retaining information across multiple interactions over time.

For example, an agent could remember a user's preferences or past interactions to provide more relevant responses in future conversations.

### **Supervision and Control Mechanisms**

Supervision and control mechanisms allow users to oversee and manage the agent's behavior. This can include permission scopes, human-in-the-loop oversight, override controls, and guardrails or policies to ensure that the agent operates within acceptable limits.

Long story short, agents are group of models that talk to each other, have memory, and can perform tasks on their own. They are a powerful new paradigm in AI that has the potential to revolutionize how we interact with technology and solve complex problems.

If we just try to connect multiple `LLMs` together to form an agent, we will run into issues with memory and coordination. We need a way to manage the interactions between the models and ensure that they can work together effectively.

This is where `LangGraph` and `LangChain` come in. They provide a framework for building agents that can manage memory, coordinate interactions between models, and perform tasks autonomously.

## **LangChain**

`LangChain` is a framework for building agents that can manage memory, coordinate interactions between models, and perform tasks autonomously. It provides modular building blocks that let you create LLM based workflows, agents, and applications.


`LangChain` contains:

**Model** : This is the core component of `LangChain`. It provides a simple interface for interacting with language models. You can use it to generate text, answer questions, and perform other language-related tasks.

**Prompts** : This component allows you to create and manage prompts for your language models. You can use it to create templates, manage variables, and generate prompts dynamically.

**Retrievers** : This component provides tools for retrieving information from external sources, such as databases or APIs. You can use it to access information that your language models may not have in their training data.

But the biggest offering of `LangChain` is `Chains`.

<hr>

### **Applications of `LangChain`**

1. **Chatbots**: You can use `LangChain` to build chatbots that can have more natural and engaging conversations with users. The memory and autonomy features of `LangChain` allow chatbots to remember past interactions and provide more personalized responses.

2. **RAG (Retrieval Augmented Generation)**: You can use `LangChain` to build RAG systems that can retrieve information from external sources and use it to generate more accurate and relevant responses.

3. **Basic Level Agents**: You can use `LangChain` to build basic level agents that can perform simple tasks autonomously, such as scheduling appointments or sending emails. `LangChain` provides `Tools` that allow agents to interact with external resources and perform tasks on their own. For example, an agent could use a `calendar API` to schedule appointments or an email API to send emails without needing explicit instructions for every action.
